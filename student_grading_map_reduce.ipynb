```python
from collections import defaultdict
```

```python
data = [
("Alice", "Math", 88),
("Bob", "Math", 75),
("Alice", "Science", 92),
("Bob", "Science", 68),
("Alice", "English", 85),
("Bob", "English", 70),
("Charlie", "Math", 95),
("Charlie", "Science", 90),
("Charlie", "English", 100)
]
```

```python
def mapper(records):
  mapped = []
  for student,subject, score in records:
    mapped.append((student,score))
  return mapped
```

```python
def shuffle_and_sort(mapped_data):
  shuffeled = defaultdict(list)
  for student, score in mapped_data:
    shuffeled[student].append(score)

  return shuffeled
```

```python
def reducer(shuffeled_data):
  results = {}
  for student, scores in shuffeled_data.items():
    avg = sum(scores) / len(scores)
    if avg >= 90:
      grade = 'A'
    elif avg >= 80:
      grade = 'B'
    elif avg >= 70:
      grade = 'C'
    elif avg >= 60:
      grade = 'D'
    else:
      grade = 'F'

    results[student] = {
        'scores': scores,
        'average': round(avg,2),
        'grade': grade
    }

  return results
```

```python
if __name__ == '__main__':
  mapped_data = mapper(data)
  shuffeled_data = shuffle_and_sort(mapped_data)
  results = reducer(shuffeled_data)

  for student, result in results.items():
    print(f"{student}: {result}")
```
